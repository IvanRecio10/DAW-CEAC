```python
from datetime import datetime
import pandas as pd
import numpy as np
import matplotlib

class Almacen():
    def __init__(self):
        self.almacen = {}
    def agregarDatosDic(self, param):
        if 'DataFrame' in locals() or 'DataFrame' in globals():
            del DataFrame
        if type(param) not in [str, int, bool, datetime]:
            if type(param) is dict:
                for valor in param.values():  
                    if type(valor) in [list, tuple, dict, set]:
                        for posicion in valor:
                            if type(posicion).__name__ in self.almacen:
                                self.almacen[type(posicion).__name__].append(posicion)
                            else:
                                self.almacen[type(posicion).__name__] = [posicion]
                    else:   
                        if type(valor).__name__ in self.almacen:
                            self.almacen[type(valor).__name__].append(valor)
                        else:
                            self.almacen[type(valor).__name__] = [valor] 
            else:
                for valor in param:
                    if type(valor).__name__ in self.almacen:
                        self.almacen[type(valor).__name__].append(valor)
                    else:
                        self.almacen[type(valor).__name__] = [valor]    
        else:
            if type(param).__name__ in self.almacen:
                self.almacen[type(param).__name__].append(param)
            else:
                self.almacen[type(param).__name__] = [param]

    def crearDataFrame(self):
        #Rellenamos con valores NaH para que el DataFrame se quede relleno de forma homogenea
        valorMax = 0
        #Sacamos el nuemro max de filas que va a tener nuestro dataFrame recoriendo los valores de nuestro diccionario
        for valor in self.almacen.values():
            if len(valor) > valorMax:
                valorMax = len(valor)
        #Una vez hemos sacado el numero maximo de filas rellenamos las otras claves del diccionario usando la libreria numpy
        for clave, valor in self.almacen.items():
            if len(valor) != valorMax:
                while len(valor)!=valorMax:
                    self.almacen[clave].append(np.nan)
        #Despues de haber rellenado el diccionario creamos el dataFrame y le ponemos un indice que se llame ID
        DataFrame = pd.DataFrame(self.almacen)
        indice = pd.Series(range(len(DataFrame)), name='ID')
        DataFrame = DataFrame.set_index(indice)        
        #Por ultimo vaciamos el diccionario
        self.almacen = {}
        self.almacen = DataFrame
        return self.almacen
    def filtrarDataFrame(self, fila=None, columna=None, manteterCambios=False):
        if fila == None:
            fila=(0, len(self.almacen))
        if columna == None:
            columna = self.almacen.keys()
        if type(fila) == tuple and len(fila)==2:
            if type(fila[0]) == int and type(fila[1]) == int:
                if manteterCambios == False:
                    return self.almacen.loc[fila[0]:fila[1], columna]
                else:
                    DataFrame = self.almacen.loc[fila[0]:fila[1], columna]
                    self.almacen = DataFrame
                    return self.almacen
            else:
                print("Ha habido un error con la fila")
        elif type(fila) == int:
            if manteterCambios == False:
                return self.almacen.loc[fila, columna]
            else:
                DataFrame = self.almacen.loc[fila, columna]
                self.almacen = DataFrame
                return self.almacen
        else:
            print("Ha habido un error con la fila")
            

                
                
```

```python

a = Almacen()
a.agregarDatosDic({"int": [1,2,3], "str": "5", "hola":[False, True], "adios": datetime.now()})
a.crearDataFrame()
a.agregarDatosDic({"int": [1,2,4,5], "str": "5", "hola":[False, True, True], "adios": datetime.now()})
a.crearDataFrame()
a.filtrarDataFrame(2, None, False)
```
